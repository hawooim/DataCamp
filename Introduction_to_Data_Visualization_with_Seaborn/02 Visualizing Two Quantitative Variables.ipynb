```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Tips dataset

```python
tips = sns.load_dataset('tips')
tips.head()
```

## scatterplot() vs. relplot()

```python
# scatterplot()
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# relplot()
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter')
plt.show()
```

## Subplots in columns

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            col='smoker')
plt.show()
```

## Subplots in rows

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            row='smoker')
plt.show()
```

## Subplots in rows and columns

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            col='smoker', row='time')
plt.show()
```

## Subgroups for days of the week

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            col='day')
plt.show()
```

## Wrapping columns

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            col='day', col_wrap=2)
plt.show()
```

## Ordering columns

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            col='day', col_wrap=2, col_order=['Thur', 'Sat', 'Fri', 'Sun'])
plt.show()
```

## Subgroups with point size

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            size='size')
plt.show()
```

## Point size and hue

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            size='size', hue='size')
plt.show()
```

## Subgroups with point style

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            hue='smoker', style='smoker')
plt.show()
```

## Changing point transparency

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', 
            alpha=0.4)
plt.show()
```

## UCI Air Quality Data Set

```python
df = pd.read_csv('AirQualityUCI.csv', delimiter=';', parse_dates=['Date'])
df = df.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
df = df.dropna()
df.head()
```

## Scatter plot

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='scatter')
plt.show()
```

## Define na values

```python
df.describe()
```

```python
na_values = dict(zip(df.describe().columns.values, [-200 for i in range(8)]))

df = pd.read_csv('AirQualityUCI.csv', delimiter=';', parse_dates=['Date'], na_values=na_values)
df = df.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
df = df.dropna()
df.head()
```

## Scatter plot after removing -200 values

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='scatter')
plt.show()
```

## Line plot (multiple observations per x-value)

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='line')  # Default area = confidence interval of the mean
plt.show()
```

## Replacing confidence interval with standard deviation

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='line', ci='sd')
plt.show()
```

## Turning off confidence interval

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='line', ci=None)
plt.show()
```

## Subgroups by time

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='line', 
            style='Time', hue='Time')
plt.show()
```

## Adding markers

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='line', 
            style='Time', hue='Time', markers=True)
plt.show()
```

## Turning off line style

```python
sns.relplot(x='Date', y='NO2(GT)', data=df, kind='line', 
            style='Time', hue='Time', markers=True, dashes=False)  # style only affects markers
plt.show()
```
