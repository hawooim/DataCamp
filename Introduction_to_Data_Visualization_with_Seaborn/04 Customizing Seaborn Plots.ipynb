## Changing the figure style
 - Figure "style" includes background and axes
 - Preset options: "white", "dark", "whitegrid", "darkgrid", "ticks"
 - sns.set_style()

```python
sns.set_style('white')
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.show()
```

```python
sns.set_style('dark')
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.show()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.show()
```

```python
sns.set_style('ticks')
sns.scatterplot(x='horsepower', y='mpg', data=mpg)
plt.show()
```

## Countries-of-the-world datasets

```python
df = pd.read_csv('countries-of-the-world.csv')
df = df.dropna()
df.head()
```

```python
df.info()
```

```python
[i for i in df.columns.values if df[i].dtype == 'O']
```

```python
target_columns = [
 'Pop. Density (per sq. mi.)',
 'Coastline (coast/area ratio)',
 'Net migration',
 'Infant mortality (per 1000 births)',
 'Literacy (%)',
 'Phones (per 1000)',
 'Arable (%)',
 'Crops (%)',
 'Other (%)',
 'Climate',
 'Birthrate',
 'Deathrate',
 'Agriculture',
 'Industry',
 'Service']
```

```python
for i in target_columns:
    df[i] = df[i].apply(lambda x: x.replace(',', '.'))
    df[i] = df[i].astype('float')
```

## Adding a title to FacetGrid

```python
# without title
g = sns.catplot(x='Region', y='Birthrate', data=df, kind='box')
plt.show()
```

```python
# add a title
g = sns.catplot(x='Region', y='Birthrate', data=df, kind='box')
g.fig.suptitle("Distribution of Birth Rate by Region")
plt.show()
```

```python
# adjusting height of title
g = sns.catplot(x='Region', y='Birthrate', data=df, kind='box')
g.fig.suptitle("Distribution of Birth Rate by Region", y=1.03)
plt.show()
```

## Adding a title to AxesSubplot

```python
# FacetGrid
g = sns.catplot(x='Region', y='Birthrate', data=df, kind='box')
g.fig.suptitle("Distribution of Birth Rate by Region", y=1.03)
plt.show()
```

```python
# AxesSubplot
g = sns.boxplot(x='Region', y='Birthrate', data=df)
g.set_title("Distribution of Birth Rate by Region", y=1.03)
plt.show()
```

## Titles for subplots

```python
df['Group'] = df['GDP ($ per capita)'] >= df['GDP ($ per capita)'].mean()
```

```python
g = sns.catplot(x='Region', y='Birthrate', data=df, kind='box', col='Group')
g.fig.suptitle("Distribution of Birth Rate by Region, between two Groups", y=1.03)
g.set_titles('This is {col_name}')
plt.show()
```

## Adding axis labels

```python
g = sns.boxplot(x='Region', y='Birthrate', data=df)
g.set_title("Distribution of Birth Rate by Region", y=1.03)
g.set(xlabel='World regions', ylabel='Birthrate per 1000')
plt.show()
```

## Rotating x-axis tick labels

```python
g = sns.boxplot(x='Region', y='Birthrate', data=df)
g.set_title("Distribution of Birth Rate by Region", y=1.03)
g.set(xlabel='World regions', ylabel='Birthrate per 1000')
plt.xticks(rotation=90)
plt.show()
```
