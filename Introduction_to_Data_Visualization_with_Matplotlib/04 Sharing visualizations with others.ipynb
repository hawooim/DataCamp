```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('climate_change.csv', parse_dates = ['date'], index_col='date')
df.head()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(df['co2'], df['relative_temp'], c=df.index)
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
plt.show()
```

## Saving the figure to file

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(df['co2'], df['relative_temp'], c=df.index)
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')

fig.savefig('scatter.png')
```

## Different file formats

```python
fig.savefig('scatter.jpg')
fig.savefig('scatter_50.jpg', quality=50)
fig.savefig('scatter_50.svg')
```

## Resolution

```python
fig.savefig('scatter_300dpi.png', dpi=300)
```

```python
df = pd.read_csv('summer2016.csv')
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

## Bar-chart of heights for all sports

```python
fig, ax = plt.subplots(figsize=(16, 8))

for sport in df['Sport'].unique():
    sport_df = df[df['Sport']==sport]
    ax.bar(sport, 
           sport_df['Height'].mean(), 
           yerr=sport_df['Height'].std())
ax.set_ylabel('Height (cm)')
ax.set_xticklabels(df['Sport'].unique(), rotation=90)
plt.show()
```
