```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('climate_change.csv', parse_dates = ['date'], index_col='date')
df.head()
```

## Plotting time-series data

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(df.index, df['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```

## Zooming in on a decade

```python
sixties = df['1960-01-01':'1969-12-31']

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(sixties.index, sixties['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```

## Zooming in on one year

```python
sixty_nine = df['1969-01-01':'1969-12-31']

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(sixty_nine.index, sixty_nine['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```

## Plotting two time-series together

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(df.index, df['co2'])
ax.plot(df.index, df['relative_temp'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm) / Relative temperature')
plt.show()
```

## Using twin axes

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(df.index, df['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')

ax2 = ax.twinx()
ax2.plot(df.index, df['relative_temp'])
ax2.set_ylabel('Relative temperature')

plt.show()
```

## Separating variables by color

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(df.index, df['co2'], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')

ax2 = ax.twinx()
ax2.plot(df.index, df['relative_temp'], color='red')
ax2.set_ylabel('Relative temperature', color='red')

plt.show()
```

## Coloring the ticks

```python
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(df.index, df['co2'], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax.tick_params('y', colors='blue')  # color's'

ax2 = ax.twinx()
ax2.plot(df.index, df['relative_temp'], color='red')
ax2.set_ylabel('Relative temperature', color='red')
ax2.tick_params('y', colors='red')  # color's'

plt.show()
```

## A function that plots time-series

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
```

## Using our function

```python
fig, ax = plt.subplots(figsize=(10, 10))
plot_timeseries(ax, df.index, df['co2'], 'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['relative_temp'], 'red', 'Time', 'Relative temperature (Celsius)')
plt.show()
```

## Annotation

```python
fig, ax = plt.subplots(figsize=(10, 10))
plot_timeseries(ax, df.index, df['co2'], 'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()
plot_timeseries(ax2, df.index, df['relative_temp'], 'red', 'Time', 'Relative temperature (Celsius)')
ax2.annotate('>1 degree', 
             xy=(pd.Timestamp('2015-10-06'), 1), 
             xytext=(pd.Timestamp('2008-10-06'), -0.2),
             arrowprops={'arrowstyle': '->', 'color': 'gray'})
plt.show()
```
