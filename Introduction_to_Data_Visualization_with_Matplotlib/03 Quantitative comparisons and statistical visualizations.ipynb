```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('medals_by_country_2016.csv', index_col=0)
df.head()
```

## Olympic medals: visualizing the data

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(df.index, df['Gold'])
plt.show()
```

## Interlude: rotate the tick labels

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(df.index, df['Gold'])
ax.set_xticklabels(df.index, rotation=90)
plt.show()
```

## Olympic medals: visualizing the other medals

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(df.index, df['Gold'])
ax.bar(df.index, df['Silver'], bottom=df['Gold'])
ax.set_xticklabels(df.index, rotation=90)
plt.show()
```

## Olympic medals: visualizing all three

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(df.index, df['Gold'])
ax.bar(df.index, df['Silver'], bottom=df['Gold'])
ax.bar(df.index, df['Bronze'], bottom=df['Gold'] + df['Silver'])
ax.set_xticklabels(df.index, rotation=90)
plt.show()
```

## Adding a legend

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar(df.index, df['Gold'], label='Gold', color='gold')
ax.bar(df.index, df['Silver'], bottom=df['Gold'], label='Silver', color='silver')
ax.bar(df.index, df['Bronze'], bottom=df['Gold'] + df['Silver'], label='Bronze', color='brown')
ax.set_xticklabels(df.index, rotation=90)
ax.set_ylabel('Number of medals')
ax.legend()
plt.show()
```

```python
df = pd.read_csv('summer2016.csv')
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
mens_rowing = df[(df['Sport']=='Rowing') & (df['Sex']=='M')]
mens_gymnastics = df[(df['Sport']=='Gymnastics') & (df['Sex']=='M')]
```

## Introducing histograms

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(mens_rowing['Height'])
ax.hist(mens_gymnastics['Height'])
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
plt.show()
```

## Labels are needed

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(mens_rowing['Height'], label='Rowing')
ax.hist(mens_gymnastics['Height'], label='Gymnastics')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

## Customizing histograms: setting the number of bins

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(mens_rowing['Height'], label='Rowing', bins=5)
ax.hist(mens_gymnastics['Height'], label='Gymnastics', bins=5)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

## Customizing histograms: setting bin boundaries

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(mens_rowing['Height'], label='Rowing', bins=list(range(150, 220, 10)))
ax.hist(mens_gymnastics['Height'], label='Gymnastics', bins=list(range(150, 220, 10)))
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

## Customizing histograms: transparency

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(mens_rowing['Height'], label='Rowing', bins=list(range(150, 220, 10)), histtype='step')
ax.hist(mens_gymnastics['Height'], label='Gymnastics', bins=list(range(150, 220, 10)), histtype='step')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

## A bar chart again

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar('Rowing', mens_rowing['Height'].mean())
ax.bar('Gymnastics', mens_gymnastics['Height'].mean())
ax.set_ylabel('Avg. Height (cm)')
plt.show()
```

## Adding error bars to bar charts

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.bar('Rowing', mens_rowing['Height'].mean(), yerr=mens_rowing['Height'].std())
ax.bar('Gymnastics', mens_gymnastics['Height'].mean(), yerr=mens_gymnastics['Height'].std())
ax.set_ylabel('Avg. Height (cm)')
plt.show()
```

```python
df = pd.read_csv('climate_change.csv', parse_dates = ['date'], index_col='date')
df.head()
```

## Introducing scatter plots

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(df['co2'], df['relative_temp'])
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
plt.show()
```

## Customizing scatter plots

```python
eighties = df['1980-01-01':'1989-12-31']
nineties = df['1990-01-01':'1999-12-31']

fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(eighties['co2'], eighties['relative_temp'], color='red', label='1980s')
ax.scatter(nineties['co2'], nineties['relative_temp'], color='blue', label='1990s')
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
ax.legend()
plt.show()
```

## Encoding a third variable by color

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(df['co2'], df['relative_temp'], c=df.index)
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
plt.show()
```
