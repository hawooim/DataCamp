```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
seattle_weather = pd.read_csv('seattle_weather.csv')
austin_weather = pd.read_csv('austin_weather.csv')
```

```python
seattle_weather['DATE'][:12]
```

```python
seattle_weather['MLY-TAVG-NORMAL'][:12]
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(seattle_weather['DATE'][:12], seattle_weather['MLY-TAVG-NORMAL'][:12])
ax.plot(austin_weather['DATE'][:12], austin_weather['MLY-TAVG-NORMAL'][:12])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(seattle_weather['DATE'][:12], seattle_weather['MLY-TAVG-NORMAL'][:12], marker='o')
ax.plot(austin_weather['DATE'][:12], austin_weather['MLY-TAVG-NORMAL'][:12], marker='v')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(seattle_weather['DATE'][:12], seattle_weather['MLY-TAVG-NORMAL'][:12], marker='o', linestyle='--')
ax.plot(austin_weather['DATE'][:12], austin_weather['MLY-TAVG-NORMAL'][:12], marker='v', linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(seattle_weather['DATE'][:12], seattle_weather['MLY-TAVG-NORMAL'][:12], marker='o', linestyle='None')
ax.plot(austin_weather['DATE'][:12], austin_weather['MLY-TAVG-NORMAL'][:12], marker='v', linestyle='None')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(seattle_weather['DATE'][:12], seattle_weather['MLY-TAVG-NORMAL'][:12], marker='o', linestyle='-')
ax.plot(austin_weather['DATE'][:12], austin_weather['MLY-TAVG-NORMAL'][:12], marker='v', linestyle='-')
ax.set_xlabel('Time (months)')
ax.set_ylabel('Average temperature (Fahrenheit degrees)')
ax.set_title('Weather in Seattle and Austin')
plt.show()
```

## Subplots with data

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
ax[0].plot(seattle_weather['DATE'][:12], seattle_weather['MLY-PRCP-NORMAL'][:12], color='blue')
ax[0].plot(seattle_weather['DATE'][:12], seattle_weather['MLY-PRCP-25PCTL'][:12], color='blue', linestyle='--')
ax[0].plot(seattle_weather['DATE'][:12], seattle_weather['MLY-PRCP-75PCTL'][:12], color='blue', linestyle='--')
ax[1].plot(austin_weather['DATE'][:12], austin_weather['MLY-PRCP-NORMAL'][:12], color='red')
ax[1].plot(austin_weather['DATE'][:12], austin_weather['MLY-PRCP-25PCTL'][:12], color='red', linestyle='--')
ax[1].plot(austin_weather['DATE'][:12], austin_weather['MLY-PRCP-75PCTL'][:12], color='red', linestyle='--')
ax[0].set_ylabel('Precipitation (inches)')
ax[1].set_ylabel('Precipitation (inches)')
ax[1].set_xlabel('Time (months)')
plt.show()
```

## Sharing the y-axis range

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharey=True)
ax[0].plot(seattle_weather['DATE'][:12], seattle_weather['MLY-PRCP-NORMAL'][:12], color='blue')
ax[0].plot(seattle_weather['DATE'][:12], seattle_weather['MLY-PRCP-25PCTL'][:12], color='blue', linestyle='--')
ax[0].plot(seattle_weather['DATE'][:12], seattle_weather['MLY-PRCP-75PCTL'][:12], color='blue', linestyle='--')
ax[1].plot(austin_weather['DATE'][:12], austin_weather['MLY-PRCP-NORMAL'][:12], color='red')
ax[1].plot(austin_weather['DATE'][:12], austin_weather['MLY-PRCP-25PCTL'][:12], color='red', linestyle='--')
ax[1].plot(austin_weather['DATE'][:12], austin_weather['MLY-PRCP-75PCTL'][:12], color='red', linestyle='--')
ax[0].set_ylabel('Precipitation (inches)')
ax[1].set_ylabel('Precipitation (inches)')
ax[1].set_xlabel('Time (months)')
plt.show()
```
