```python
# import all modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
mpg = sns.load_dataset('mpg')
```

## Pandas histogram
 - Actual frequency of observations
 - No automatic labels
 - Wide bins

```python
mpg['weight'].plot.hist()
plt.show()
```

## Seaborn distplot
 - Automatic label on x-axis
 - Muted color palette
 - KDE plot (Gaussian Kernel Density Estimate)
 - Narrow bins

```python
sns.distplot(mpg['weight'])
plt.show()
```

```python
g = sns.displot(mpg['weight'])
plt.show()
type(g)
```

```python
g = sns.histplot(mpg['weight'])
plt.show()
type(g)
```

## Creating a histogram using sns.distplot()

```python
sns.distplot(mpg['weight'], kde=False, bins=10)
plt.show()
```

## Alternative data distributions

```python
sns.distplot(mpg['weight'], hist=False, rug=True)
plt.show()
```

## Further customization

```python
sns.distplot(mpg['weight'], hist=False, rug=True, kde_kws={'shade':True})
plt.show()
```

## Introduction to regplot

```python
g = sns.regplot(x='weight', y='mpg', data=mpg)
plt.show()
type(g)
```

## lmplot() builds on top of the base regplot()

```python
g = sns.lmplot(x='weight', y='mpg', data=mpg)
plt.show()
type(g)
```

## lmplot() faceting

```python
# Organize data by colors (hue)
g = sns.lmplot(x='weight', y='mpg', data=mpg, hue='origin')
plt.show()
```

```python
# Organize data by columns (col)
g = sns.lmplot(x='weight', y='mpg', data=mpg, col='origin')
plt.show()
```

```python
# Organize data by rows (row)
g = sns.lmplot(x='weight', y='mpg', data=mpg, row='origin')
plt.show()
```
